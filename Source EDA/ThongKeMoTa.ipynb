```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
df = pd.read_csv("DATA/wage_eda.csv")

# ── Overview ──────────────────────────────────────────────────
print(f"Shape: {df.shape[0]} rows × {df.shape[1]} columns")
print(f"Duplicate rows: {df.duplicated().sum()}")

# ── Missing values ────────────────────────────────────────────
missing = df.isnull().sum()
print("\nMissing Values")
display(missing[missing > 0].to_frame("count").assign(pct=lambda x: (x["count"]/len(df)*100).round(2)))

# ── Descriptive stats – continuous variables ──────────────────
continuous = ["lwage", "wage", "educ", "exper", "IQ", "KWW", "fatheduc", "motheduc"]
print("\nDescriptive Statistics – Continuous Variables")
display(df[continuous].describe().T.round(3))

# ── Binary variables summary (proportion = 1) ─────────────────
binary = ["black", "south", "smsa", "nearc2", "nearc4",
          "momdad14", "sinmom14", "step14", "south66", "smsa66", "libcrd14"]
print("\nBinary Variables – Proportion = 1")
summary = pd.DataFrame({
    "N (=1)": df[binary].sum().astype(int),
    "N (=0)": (df[binary] == 0).sum().astype(int),
    "% (=1)": (df[binary].mean() * 100).round(2)
})
display(summary)

# ── married ───────────────────────────────────────────────────
married_map = {1:"Married", 2:"Widowed", 3:"Divorced",
               4:"Separated", 5:"Never married", 6:"Other"}
print("\nMarital Status Distribution")
m = df["married"].map(married_map).value_counts()
display(m.to_frame("count").assign(pct=(m/len(df)*100).round(2)))

# ── Region dummies → single region label ─────────────────────
reg_cols = ["reg661","reg662","reg663","reg664","reg665","reg666","reg667","reg668","reg669"]
reg_map  = {"reg661":"New England","reg662":"Mid Atlantic","reg663":"E N Central",
            "reg664":"W N Central","reg665":"South Atlantic","reg666":"E S Central",
            "reg667":"W S Central","reg668":"Mountain","reg669":"Pacific"}
df["region"] = df[reg_cols].idxmax(axis=1).map(reg_map)
print("\nRegion Distribution")
r = df["region"].value_counts()
display(r.to_frame("count").assign(pct=(r/len(df)*100).round(2)))

# ── Skewness & Kurtosis ───────────────────────────────────────
print("\nSkewness & Kurtosis – Continuous Variables")
display(df[continuous].agg(["skew","kurt"]).T.round(3))
```
